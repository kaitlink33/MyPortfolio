## Problem 3

Ask the user for the radius of a circle (decimal allowed). Calculate and print the area rounded to **2 decimal places**.  

The formula is:  $Area = \pi \times r^2$

consider pi = 3.14159 

`Enter the radius of the circle: 5`

`The area of the circle is 78.54`

```python
radius = float(input("What is the radius of a circle?"))
area = (radius * radius) * 3.14159
print(f"The area of the circle is {area:,.2f}")
```

## Problem 6
Write a short code snippet that takes in the **name of a movie** and its **ticket price as a decimal number**. Then you should output how much it would cost to buy 2, 3, 5, and 8 tickets for that movie. The amounts should be formatted as currency, with commas and two decimals. You must use **f-string**. To format a number with commas using f-strings, 

Example output for a movie *Demon Slayer* that costs 9.75 per ticket would look like:

`A 'Demon Slayer' ticket costs $9.75.`<br/>  
&nbsp;&nbsp;&nbsp;&nbsp;`2   tickets cost $19.50`<br/>  
&nbsp;&nbsp;&nbsp;&nbsp;`3   tickets cost $29.25`<br/>  
&nbsp;&nbsp;&nbsp;&nbsp;`5   tickets cost $48.75`<br/>  
&nbsp;&nbsp;&nbsp;&nbsp;`8   tickets cost $78.00`<br/>  

*Note that it is easier to finish one of the cost lines first, then copy/paste and modify for the rest.*


```python
# code here
movie_name=input("What is the name of the movie?")
ticket_price=float(input("How much do the tickets cost?"))
print(f"A '{movie_name}' ticket costs ${ticket_price:,.2f}.")
print(f"2 tickets cost ${2*ticket_price:,.2f}")
print(f"3 tickets cost ${3*ticket_price:,.2f}")
print(f"5 tickets cost ${5*ticket_price:,.2f}")
print(f"8 tickets cost ${8*ticket_price:,.2f}")

```
